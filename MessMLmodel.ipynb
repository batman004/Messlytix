```python


import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pickle
```

```python

data = pd.read_csv(r'DATASET.csv')
data.head()

```

```python
data['Menu Rating'].fillna(data['Menu Rating'].mean(), inplace=True)
data['Amount Of Food Cooked'].fillna(data['Amount Of Food Cooked'].mean(), inplace=True)
data['Wastage'].fillna(data['Wastage'].mean(), inplace=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder as LE
```

```python
#For days

s=data['Day']
 
data.apply(lambda s: s.map({k:i for i,k in enumerate(s.unique())}))

```

```python
predict = "Amount Of Food Cooked"
X = np.array(data.drop([predict], 1))
y = np.array(data[predict])

X
```

```python
# TRAIN MODEL MULTIPLE TIMES FOR BEST SCORE
best = 0
limit=97.0
x=0.0

for _ in range(200):
    x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    X, y, test_size=0.3)

    linear = linear_model.LinearRegression()

    linear.fit(x_train, y_train)
    acc = linear.score(x_test, y_test)
    print("Accuracy: " + str(acc))

    if acc > best:
        best = acc
        with open("food.pickle", "wb") as f:
            pickle.dump(linear, f)

```

```python
pickle_in = open("food.pickle", "rb")
linear = pickle.load(pickle_in)
acc
```

```python
predicted= linear.predict(x_test)
x_test1=np.array(x_test)
y_test1=np.array(y_test)
```

```python
#Comparing predicitons with actual results
print("Actual Result                       Input Params                        Predicted Result")
print()
for w in range(len(y_test)):
    print(y_test1[w],'',x_test1[w],'     ',"{:.2f}".format(predicted[w]))
    print()
```
